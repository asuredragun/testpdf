````python
import os
import json
import pandas as pd
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.document_loaders import PyPDFLoader
from dotenv import load_dotenv

# Cargar variables del .env
load_dotenv()


# Definir el modelo LLM
api_key = os.getenv("API_KEY")
llm = ChatOpenAI(model="gpt-4o-mini", openai_api_key=api_key)

# Crear un prompt para procesar tablas
prompt = PromptTemplate(
    input_variables=["pdf_content"],
    template=(
        "Extrae las tablas del siguiente texto del PDF y organ√≠zalas en un formato tabular JSON."
        "El texto es:\n{pdf_content}."
        "Solo devuelve el JSON, sin ```json\n"
    )
)

# Crear una cadena LangChain
chain = LLMChain(llm=llm, prompt=prompt)

# Directorio de entrada y salida
pdf_directory = "pdf"
output_directory = "json_output"

# Crear directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Procesar PDFs en el directorio
for filename in os.listdir(pdf_directory):
    if filename.endswith(".pdf"):
        filepath = os.path.join(pdf_directory, filename)
        output_filepath = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}.json")
        
        print(f"üìÑ Procesando archivo: {filename}")

        try:
            # Cargar y leer el PDF
            loader = PyPDFLoader(filepath)
            documents = loader.load()
            pdf_text = " ".join([doc.page_content for doc in documents])

            # Ejecutar el modelo en el contenido del PDF
            response = chain.run(pdf_content=pdf_text)
            data = json.loads(response)  # Convertir respuesta a JSON
            
            # Guardar el resultado en un JSON separado
            with open(output_filepath, "w", encoding="utf-8") as f:
                json.dump(data, f, indent=4, ensure_ascii=False)
            
            print(f"‚úÖ Archivo guardado: {output_filepath}")

        except Exception as e:
            print(f"‚ùå Error al procesar {filename}: {e}")

print("üéâ Procesamiento completado. Todos los JSON fueron generados en 'json_output'.")


````

````python
import os
import json
import pandas as pd

# Directorio donde est√°n los JSON
json_directory = "json_output"

# Funci√≥n para convertir JSON en DataFrame manejando estructuras complejas
def json_to_dataframe(data, folio=None):
    if isinstance(data, list):
        df = pd.json_normalize(data)
    elif isinstance(data, dict):
        df = pd.json_normalize([data])
    else:
        return pd.DataFrame()  # Si es nulo o inv√°lido, devolver DataFrame vac√≠o

    # Asegurar que cada entrada tenga el folio correcto para mantener la relaci√≥n
    if folio:
        df["folio"] = folio

    # Reemplazar valores faltantes con None (equivalente a null en JSON)
    df = df.where(pd.notna(df), None)
    
    return df

# Inicializar DataFrames vac√≠os para almacenar los resultados combinados
df_main_combined = pd.DataFrame()
df_residuos_combined = pd.DataFrame()
df_comprobantes_combined = pd.DataFrame()
df_responsables_combined = pd.DataFrame()

# Procesar cada archivo JSON en el directorio
for filename in os.listdir(json_directory):
    if filename.endswith(".json"):
        filepath = os.path.join(json_directory, filename)

        try:
            # Cargar el archivo JSON
            with open(filepath, "r", encoding="utf-8") as f:
                data = json.load(f)

            # Obtener el folio del JSON (asegurarse de que haya un identificador √∫nico)
            folio = data.pop("n_folio", None) or data.get("folio", filename)

            # Extraer datos anidados (si no existen, asignar lista vac√≠a para evitar errores)
            residuos = data.pop("residuos", [])
            comprobantes = data.pop("comprobante", [])
            responsables = data.pop("responsable", [])

            # Convertir datos principales en DataFrame
            df_main = json_to_dataframe(data, folio)

            # Convertir estructuras anidadas a DataFrames, manteniendo la relaci√≥n con el folio
            df_residuos = json_to_dataframe(residuos, folio)
            df_comprobantes = json_to_dataframe(comprobantes, folio)
            df_responsables = json_to_dataframe(responsables, folio)

            # Asegurar que los valores nulos sean None y no se llenen incorrectamente
            df_main = df_main.where(pd.notna(df_main), None)
            df_residuos = df_residuos.where(pd.notna(df_residuos), None)
            df_comprobantes = df_comprobantes.where(pd.notna(df_comprobantes), None)
            df_responsables = df_responsables.where(pd.notna(df_responsables), None)

            # Concatenar los DataFrames de forma segura
            df_main_combined = pd.concat([df_main_combined, df_main], ignore_index=True)
            df_residuos_combined = pd.concat([df_residuos_combined, df_residuos], ignore_index=True)
            df_comprobantes_combined = pd.concat([df_comprobantes_combined, df_comprobantes], ignore_index=True)
            df_responsables_combined = pd.concat([df_responsables_combined, df_responsables], ignore_index=True)

        except Exception as e:
            print(f"‚ùå Error al procesar {filename}: {e}")

# Mostrar las tablas combinadas en pantalla
print("üìù Datos principales combinados:")
display(df_main_combined)

print("üìù Residuos combinados:")
display(df_residuos_combined)

print("üìù Comprobantes combinados:")
display(df_comprobantes_combined)

print("üìù Responsables combinados:")
display(df_responsables_combined)

print("üéâ Conversi√≥n completada.")


````

````python
import pandas as pd

# Funci√≥n para convertir el n√∫mero de columna a su equivalente en letras
def colnum_to_excel(colnum):
    result = ""
    while colnum >= 0:
        result = chr(colnum % 26 + 65) + result
        colnum = colnum // 26 - 1
    return result

# Definir el archivo de salida
output_excel_filepath = "output_combined_data.xlsx"

# Guardar los DataFrames combinados en un archivo Excel con tres hojas
with pd.ExcelWriter(output_excel_filepath, engine="xlsxwriter") as writer:
    # Guardar cada DataFrame en una hoja distinta
    df_main_combined.to_excel(writer, index=False, sheet_name="Datos Principales")
    df_residuos_combined.to_excel(writer, index=False, sheet_name="Residuos")
    df_comprobantes_combined.to_excel(writer, index=False, sheet_name="Comprobantes")
    
    # Acceder a las hojas para aplicar formato de tabla
    workbook  = writer.book
    worksheet_main = writer.sheets["Datos Principales"]
    worksheet_residuos = writer.sheets["Residuos"]
    worksheet_comprobantes = writer.sheets["Comprobantes"]
    
    # Definir el rango de celdas para cada tabla (suponiendo que no hay encabezados vac√≠os)
    main_end_col = colnum_to_excel(df_main_combined.shape[1] - 1)
    residuos_end_col = colnum_to_excel(df_residuos_combined.shape[1] - 1)
    comprobantes_end_col = colnum_to_excel(df_comprobantes_combined.shape[1] - 1)
    
    worksheet_main.add_table(f"A1:{main_end_col}{df_main_combined.shape[0] + 1}")
    worksheet_residuos.add_table(f"A1:{residuos_end_col}{df_residuos_combined.shape[0] + 1}")
    worksheet_comprobantes.add_table(f"A1:{comprobantes_end_col}{df_comprobantes_combined.shape[0] + 1}")

print(f"‚úÖ Archivo Excel guardado como: {output_excel_filepath}")

````
