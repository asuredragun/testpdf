````python
import os
import json
import pandas as pd
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.document_loaders import PyPDFLoader
from dotenv import load_dotenv

# Cargar variables del .env
load_dotenv()


# Definir el modelo LLM
api_key = os.getenv("API_KEY")
llm = ChatOpenAI(model="gpt-4o-mini", openai_api_key=api_key)

# Crear un prompt para procesar tablas
prompt = PromptTemplate(
    input_variables=["pdf_content"],
    template=(
        "Extrae las tablas del siguiente texto del PDF y organízalas en un formato tabular JSON."
        "El texto es:\n{pdf_content}."
        "Solo devuelve el JSON, sin ```json\n"
    )
)

# Crear una cadena LangChain
chain = LLMChain(llm=llm, prompt=prompt)

# Directorio de entrada y salida
pdf_directory = "pdf"
output_directory = "json_output"

# Crear directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Procesar PDFs en el directorio
for filename in os.listdir(pdf_directory):
    if filename.endswith(".pdf"):
        filepath = os.path.join(pdf_directory, filename)
        output_filepath = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}.json")
        
        print(f"📄 Procesando archivo: {filename}")

        try:
            # Cargar y leer el PDF
            loader = PyPDFLoader(filepath)
            documents = loader.load()
            pdf_text = " ".join([doc.page_content for doc in documents])

            # Ejecutar el modelo en el contenido del PDF
            response = chain.run(pdf_content=pdf_text)
            data = json.loads(response)  # Convertir respuesta a JSON
            
            # Guardar el resultado en un JSON separado
            with open(output_filepath, "w", encoding="utf-8") as f:
                json.dump(data, f, indent=4, ensure_ascii=False)
            
            print(f"✅ Archivo guardado: {output_filepath}")

        except Exception as e:
            print(f"❌ Error al procesar {filename}: {e}")

print("🎉 Procesamiento completado. Todos los JSON fueron generados en 'json_output'.")


````

````python
import os
import json
import pandas as pd

# Directorios de entrada
json_directory = "json_output"  # Carpeta donde están los JSON

# Función para convertir JSON en DataFrame manejando estructuras complejas
def json_to_dataframe(data):
    if isinstance(data, list):
        return pd.json_normalize(data)  # Si es una lista de diccionarios, normalizar
    elif isinstance(data, dict):
        return pd.json_normalize([data])  # Si es un diccionario, meterlo en una lista antes de normalizar
    else:
        raise ValueError("Formato de JSON no compatible")

# Inicializar listas para almacenar los DataFrames combinados
df_main_combined = pd.DataFrame()  # Para los datos principales
df_residuos_combined = pd.DataFrame()  # Para los residuos
df_comprobantes_combined = pd.DataFrame()  # Para los comprobantes
df_responsables_combined = pd.DataFrame()  # Para los responsables

# Procesar cada archivo JSON en el directorio
for filename in os.listdir(json_directory):
    if filename.endswith(".json"):
        filepath = os.path.join(json_directory, filename)

        try:
            # Cargar el archivo JSON
            with open(filepath, "r", encoding="utf-8") as f:
                data = json.load(f)

            # Extraer residuos y comprobantes si existen
            residuos = data.pop("residuos", None)  # Elimina "residuos" del JSON y lo guarda aparte
            comprobantes = data.pop("comprobante", None)  # Extrae "comprobantes" del JSON
            responsables = data.pop("responsable", None)  # Extrae "responsable" del JSON

            # Convertir los datos principales a DataFrame
            df_main = json_to_dataframe(data)

            # Si hay residuos, convertirlos en otro DataFrame
            df_residuos = json_to_dataframe(residuos) if residuos else pd.DataFrame()

            # Si hay comprobantes, convertirlos en otro DataFrame
            df_comprobantes = json_to_dataframe(comprobantes) if comprobantes else pd.DataFrame()
            
            # Si hay responsable, convertirlo en otro DataFrame
            df_responsable = json_to_dataframe(responsables) if responsables else pd.DataFrame()

            # Concatenar los DataFrames en sus respectivas tablas combinadas
            df_main_combined = pd.concat([df_main_combined, df_main], ignore_index=True)
            df_residuos_combined = pd.concat([df_residuos_combined, df_residuos], ignore_index=True)
            df_comprobantes_combined = pd.concat([df_comprobantes_combined, df_comprobantes], ignore_index=True)
            df_responsables_combined = pd.concat([df_responsables_combined, df_responsable], ignore_index=True)

            # Mostrar los DataFrames en pantalla para cada archivo
            print(f"📝 Datos de {filename} - Datos principales:")
            display(df_main)
            
            if not df_residuos.empty:
                print(f"📝 Datos de {filename} - Residuos:")
                display(df_residuos)
            if not df_comprobantes.empty:
                print(f"📝 Datos de {filename} - Comprobantes:")
                display(df_comprobantes)
            if not df_responsable.empty:
                print(f"📝 Datos de {filename} - Responsable:")
                display(df_responsable)

        except Exception as e:
            print(f"❌ Error al procesar {filename}: {e}")

# Mostrar las tablas combinadas en pantalla
print("📝 Datos principales combinados:")
display(df_main_combined)

print("📝 Residuos combinados:")
display(df_residuos_combined)

print("📝 Comprobantes combinados:")
display(df_comprobantes_combined)

print("📝 Responsables combinados:")
display(df_responsables_combined)

print("🎉 Conversión completada.")



````

````python
import pandas as pd

# Función para convertir el número de columna a su equivalente en letras
def colnum_to_excel(colnum):
    result = ""
    while colnum >= 0:
        result = chr(colnum % 26 + 65) + result
        colnum = colnum // 26 - 1
    return result

# Definir el archivo de salida
output_excel_filepath = "output_combined_data.xlsx"

# Guardar los DataFrames combinados en un archivo Excel con tres hojas
with pd.ExcelWriter(output_excel_filepath, engine="xlsxwriter") as writer:
    # Guardar cada DataFrame en una hoja distinta
    df_main_combined.to_excel(writer, index=False, sheet_name="Datos Principales")
    df_residuos_combined.to_excel(writer, index=False, sheet_name="Residuos")
    df_comprobantes_combined.to_excel(writer, index=False, sheet_name="Comprobantes")
    
    # Acceder a las hojas para aplicar formato de tabla
    workbook  = writer.book
    worksheet_main = writer.sheets["Datos Principales"]
    worksheet_residuos = writer.sheets["Residuos"]
    worksheet_comprobantes = writer.sheets["Comprobantes"]
    
    # Definir el rango de celdas para cada tabla (suponiendo que no hay encabezados vacíos)
    main_end_col = colnum_to_excel(df_main_combined.shape[1] - 1)
    residuos_end_col = colnum_to_excel(df_residuos_combined.shape[1] - 1)
    comprobantes_end_col = colnum_to_excel(df_comprobantes_combined.shape[1] - 1)
    
    worksheet_main.add_table(f"A1:{main_end_col}{df_main_combined.shape[0] + 1}")
    worksheet_residuos.add_table(f"A1:{residuos_end_col}{df_residuos_combined.shape[0] + 1}")
    worksheet_comprobantes.add_table(f"A1:{comprobantes_end_col}{df_comprobantes_combined.shape[0] + 1}")

print(f"✅ Archivo Excel guardado como: {output_excel_filepath}")

````
